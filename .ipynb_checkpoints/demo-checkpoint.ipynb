```python
def average_anomalies(df, avg3days =True):
  rescp = df.copy()
  for index, row in rescp.iterrows():
    if(row['isAnomaly']):
      if(avg3days):
          for i in range(3):
              try:
                rescp.at[index + i - 1,'value'] = (rescp.loc[index - 3 + i -1]['value'] + rescp.loc[index+3 +i -1]['value'])/2
              except:
                try:
                  rescp.at[index + i - 1,'value'] = rescp.loc[index - 3 + i -1]['value']
                except:
                  rescp.at[index + i - 1,'value'] = rescp.loc[index+3 +i -1]['value']
      else:
        try:
          rescp.at[index,'value'] = (rescp.loc[index - 3]['value'] + rescp.loc[index+3]['value'])/2
        except:
          try:
            rescp.at[index,'value'] = rescp.loc[index+3]['value']
          except:
            rescp.at[index,'value'] = rescp.loc[index - 3]['value'] 
  return rescp
```

```python
def decompose(df):
    rescp = df.copy()
    from statsmodels.tsa.seasonal import STL
    vincom = pd.Series(rescp['value'].values,
                    index=pd.date_range(rescp['timestamp'].min(),
                                        periods=len(rescp['value'].values),
                                        freq='d'),
                    name = 'time')
    stl = STL(vincom, seasonal=7)
    res = stl.fit()
    return (res.trend, res.seasonal, res.resid, res)
```

```python
#convert series to supervised data | use for LSTM
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    dff = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(dff.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(dff.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
#predict residual using LSTM
def predict_resid(resid):
    dfresid = pd.DataFrame(data ={'value': resid})

    values = dfresid.values 
    # ensure all data is float
    values = values.astype('float32')
    # normalize features
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(values)
    # frame as supervised learning
    reframed = series_to_supervised(scaled, STEP, STEP)

    # split into train and test sets
    values = reframed.values

    n_train_time = int (len(values)-STEP)
    train = values[:n_train_time, :]
    test = values[n_train_time:, :]
    train_X, train_y = train[:, :-STEP], train[:, -STEP:]
    test_X, test_y = test[-1:, -STEP:], test[:, -STEP:]
    train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
    test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))


    #Network design
    model = Sequential()
    model.add(LSTM(100, input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences = True))
    model.add(Dropout(0.2))
    model.add(LSTM(70))
    model.add(Dropout(0.2))
    model.add(Dense(STEP))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # fit network
    # history = model.fit(train_X, train_y, epochs=20, batch_size=1, validation_data=(test_X, test_y), verbose=2, shuffle=False)
    history = model.fit(train_X, train_y, epochs=num_epochs, batch_size=0, verbose=0, shuffle=False)

    yhat = model.predict(test_X)
    inv_yhat = scaler.inverse_transform(yhat[0].reshape(len(yhat[0]), 1))
#     inv_ytest = scaler.inverse_transform(test_y[0].reshape(len(test_y[0]), 1))

    return inv_yhat.flatten()
```

```python
threshold = 0
def mape(actual, pred):
    asp = [(abs(actual[i] - pred[i])/actual[i])*100 if actual[i] > threshold else 0 for i in range(len(pred))]
    return np.mean(np.array(asp))

def smape(actual, pred):
    asp = [(abs(actual[i] - pred[i])/(abs(actual[i]) + abs(pred[i])))*100 if actual[i] > threshold else 0 for i in range(len(pred))]
    return np.mean(np.array(asp))
```
